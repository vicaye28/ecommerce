```python
import pandas as pd 
```

```python
pd.set_option("display.max_columns", 30)
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
df= pd.read_csv('women_clothing_ecommerce_sales.csv')

df
```

```python
df.parse_dates=["order_date"]
```

```python
df
```

Converting date from a string to datetime

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d %H:%M:%S.%f)
```

```python
df.info()
```

creating month and year columns 

```python
df['month']= df['order_date'].dt.month
df['year']=df['order_date'].dt.year
```

```python
df
```

finding missing values in size column

```python
df['size'].isna().sum()
```

inputting values in place of missing values 

```python
df['size'].mode()
```

```python
df['size'].fillna('XL', inplace=True)

df['size'].value_counts()
```

```python
df['size'].replace('One SIze', 'One Size', inplace = True)


df['size'].value_counts()
```

Creating a day column too

```python
df['day'] = df['order_date'].dt.day

df['day']
```

export as csv file

```python
df.to_csv('womens_clothing_sales.csv', index=False, date_format='%Y-%m-%d %H:%M:%S')
```
